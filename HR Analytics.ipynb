# One-Hot-Encoding and Label-Encoder.

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
label_encode_columns =['salary','Role']

def encode_catergories(df,cols):
    for i in cols:
        df["cat_"+i]=labelencoder.fit_transform(df[i])
    return df
```

```python
data = encode_catergories(data,label_encode_columns)
```

```python
role_dict = {'VP':'Superior_level','Senior Director':'Superior_level','Director':'Superior_level','Senior Manager':'Higher_level','Manager':'Higher_level','Level 2-4':'Lower_level','Level 1':'Lower_level'}
```

```python
data["Role"] = data["Role"].map(role_dict)
```

```python
data["Gender"] = data["Gender"].map({'M':1,'F':0})
```

```python
data["salary"] = data["salary"].map({'low':1,'medium':2,'high':3})
```

```python
# One Hot encoding for GEO and department
from sklearn.preprocessing import OneHotEncoder
# creating instance of one-hot-encoder
encoder = OneHotEncoder(handle_unknown='ignore')
def one_hot_encoder(df, categorical_columns=None, nan_as_category=False,prefix=None):
    """Create a new column for each categorical value in categorical columns. """
    original_columns = list(df.columns)
    if not categorical_columns:
        categorical_columns = [col for col in df.columns if df[col].dtype == 'object']
    df = pd.get_dummies(df, columns=categorical_columns, dummy_na=nan_as_category,prefix=prefix)
    categorical_columns = [c for c in df.columns if c not in original_columns]
    return df, categorical_columns

data =data.join(one_hot_encoder(data[['GEO','Department']])[0])
final_data = data.drop(columns=['ID','Name','Department','GEO','Role'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
#split data into train and test
def split_train_test(df,target_col):
    X=df.drop(columns=target_col)
    y= df[target_col]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
    return X_train, X_test, y_train, y_test
```

```python
#call split_data func
X_train, X_test, y_train, y_test = split_train_test(df,'left_Company')
```

```python
X_train.shape,y_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
def apply_models(model_name,X_train,y_train,X_test,y_test):
    if model_name=="LogisticRegression":
        print(f"Below are the metrics for {model_name} model :")
        reg_log = LogisticRegression()
        reg_log.fit(X_train, y_train)
        y_pred = reg_log.predict(X_test)
        #print(metrics.classification_report(y_test, y_pred))
        print ("accuracy",metrics.accuracy_score(y_test, y_pred))
        print("roc_auc_score: ", metrics.roc_auc_score(y_test, y_pred))
        print("f1 score: ", metrics.f1_score(y_test, y_pred))
        accuracy = metrics.accuracy_score(y_test, y_pred)
        return y_pred,accuracy
    elif model_name=="RandomForest":
        print(f"Below are the metrics for {model_name} model :")
        reg_rf = RandomForestClassifier()
        reg_rf.fit(X_train, y_train)
        y_pred = reg_rf.predict(X_test)
        print(metrics.classification_report(y_test, y_pred))
        print ("accuracy",metrics.accuracy_score(y_test, y_pred))
        print("roc_auc_score: ", metrics.roc_auc_score(y_test, y_pred))
        print("f1 score: ", metrics.f1_score(y_test, y_pred))
        accuracy = metrics.accuracy_score(y_test, y_pred)
        return y_pred,accuracy
    elif model_name=="knn":
        print(f"Below are the metrics for {model_name} model :")
        reg_knn = KNeighborsClassifier()
        reg_knn.fit(X_train, y_train)
        y_pred = reg_knn.predict(X_test)
        print ("accuracy",metrics.accuracy_score(y_test, y_pred))
        #print(metrics.classification_report(y_test, y_pred))
        print("roc_auc_score: ", metrics.roc_auc_score(y_test, y_pred))
        print("f1 score: ", metrics.f1_score(y_test, y_pred))
        accuracy = metrics.accuracy_score(y_test, y_pred)
        return y_pred,accuracy
        
```

```python
execute_models =['LogisticRegression','RandomForest','knn']
dict_model ={}
for i in execute_models:
    pred,acc = apply_models(i,X_train,y_train,X_test,y_test)
    dict_model[i]=(acc,pred)


    
```
